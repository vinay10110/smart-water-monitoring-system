```python
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import StandardScaler

from pycaret.regression import *
```

# 1. Data Loading

```python
train_df=pd.read_csv('./dataset/train.csv')
test_df=pd.read_csv('./dataset/test.csv')
```

# 2. Data Preprocessing & EDA

```python
train_df=train_df.drop(columns=['Timestamp'],axis=1)
test_df=test_df.drop(columns=['Timestamp'],axis=1)
```

```python
train_df.isna().sum()
```

```python
test_df.isna().sum()
```

```python
# For train data
train_df['Apartment_Type'] = train_df['Apartment_Type'].fillna('Missing')
train_df['Amenities'] = train_df['Amenities'].fillna('Missing')

# For test data
test_df['Apartment_Type'] = test_df['Apartment_Type'].fillna('Missing')
test_df['Amenities'] = test_df['Amenities'].fillna('Missing')
```

```python
def convert_to_float_or_nan(value):
    """Converts a value to float if it's a numerical string,
    otherwise returns NaN."""
    try:
        # Check if the value matches a numerical pattern
        if re.fullmatch(r'\d+(\.\d+)?', value):
            return float(value)
        else:
            return np.nan  # Non-numerical, replace with NaN
    except (ValueError, TypeError):
        return np.nan  # Handle errors (e.g., if value is None)

# Apply the function to the 'Humidity' column
train_df['Humidity'] = train_df['Humidity'].apply(convert_to_float_or_nan)
test_df['Humidity'] = test_df['Humidity'].apply(convert_to_float_or_nan)
```

```python
# Compute medians from training data
temp_median = train_df['Temperature'].median()
humidity_median = train_df['Humidity'].median()
usage_median = train_df['Appliance_Usage'].median()

# Apply to train
train_df['Temperature'] = train_df['Temperature'].fillna(temp_median)
train_df['Humidity'] = train_df['Humidity'].fillna(humidity_median)
train_df['Appliance_Usage'] = train_df['Appliance_Usage'].fillna(usage_median)

# Apply same medians to test
test_df['Temperature'] = test_df['Temperature'].fillna(temp_median)
test_df['Humidity'] = test_df['Humidity'].fillna(humidity_median)
test_df['Appliance_Usage'] = test_df['Appliance_Usage'].fillna(usage_median)
```

```python
train_df
```

```python
valid_categories = ['Low', 'Middle', 'Upper Middle', 'Rich']

def clean_income(df):
    # Step 1: Fill NaN with 'Missing'
    df['Income_Level'] = df['Income_Level'].fillna('Missing')
    
    # Step 2: Replace invalid categories with 'Unknown'
    df.loc[~df['Income_Level'].isin(valid_categories + ['Missing']), 'Income_Level'] = 'Unknown'
    return df

# Apply to both train and test
train_df = clean_income(train_df)
test_df = clean_income(test_df)
```

```python
train_df.dtypes
```

```python
# For train_df
train_df['Apartment_Type'] = train_df['Apartment_Type'].astype('category')
train_df['Apartment_Type'] = train_df['Apartment_Type'].cat.codes

train_df['Income_Level'] = train_df['Income_Level'].astype('category')
train_df['Income_Level'] = train_df['Income_Level'].cat.codes

train_df['Amenities'] = train_df['Amenities'].astype('category')
train_df['Amenities'] = train_df['Amenities'].cat.codes

# For test_df
test_df['Apartment_Type'] = test_df['Apartment_Type'].astype('category')
test_df['Apartment_Type'] = test_df['Apartment_Type'].cat.codes

test_df['Income_Level'] = test_df['Income_Level'].astype('category')
test_df['Income_Level'] = test_df['Income_Level'].cat.codes

test_df['Amenities'] = test_df['Amenities'].astype('category')
test_df['Amenities'] = test_df['Amenities'].cat.codes
```

```python
# Columns to scale
num_cols = ['Residents', 'Temperature', 'Humidity', 'Water_Price', 
            'Period_Consumption_Index', 'Appliance_Usage']

scaler = StandardScaler()

# Fit on train
train_df[num_cols] = scaler.fit_transform(train_df[num_cols])

# Transform test using the same scaler
test_df[num_cols] = scaler.transform(test_df[num_cols])
```
